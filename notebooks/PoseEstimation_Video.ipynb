# Dibujar la pose en cada frame

```python
def drawPose(img_copy, pose_landmarks):
    size_points = 2
    naranja =(128, 0, 255)
    morado = (255, 191, 0)
    verde = (6, 243, 161)
    amarillo = (236, 243, 6) 
    rosa = (224, 18, 145)
    azul = (18, 132, 224)
    negro = (0,0,0)
    blanco = (255, 255, 255)


    height, width, _ = img_copy.shape
    
    if pose_landmarks:

        # brazo derecho
        x1 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].x * width)
        y1 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y * height)
        x2 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].x * width)
        y2 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].y * height)
        x3 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].x * width)
        y3 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].y * height)
        cv2.line(img_copy, (x1, y1), (x2, y2), (255, 255, 255), size_points)
        cv2.line(img_copy, (x2, y2), (x3, y3), (255, 255, 255), size_points)
        cv2.circle(img_copy, (x1, y1), size_points,naranja , -1)
        cv2.circle(img_copy, (x2, y2), size_points, naranja, -1)
        cv2.circle(img_copy, (x3, y3), size_points, naranja, -1)

        # brazo izquierdo
        x4 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x * width)
        y4 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y * height)
        x5 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].x * width)
        y5 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].y * height)
        x6 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].x * width)
        y6 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].y * height)
        cv2.line(img_copy, (x4, y4), (x5, y5), (255, 255, 255), 3)
        cv2.line(img_copy, (x5, y5), (x6, y6), (255, 255, 255), 3)
        cv2.circle(img_copy, (x4, y4), size_points, morado, -1)
        cv2.circle(img_copy, (x5, y5), size_points, morado, -1)
        cv2.circle(img_copy, (x6, y6), size_points, morado, -1)

        # pierna derecha
        x7 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].x * width)
        y7 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].y * height)
        x8 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].x * width)
        y8 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].y * height)
        x9 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ANKLE].x * width)
        y9 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ANKLE].y * height)
        cv2.line(img_copy, (x7, y7), (x8, y8), (255, 255, 255), size_points)
        cv2.line(img_copy, (x8, y8), (x9, y9), (255, 255, 255), size_points)
        cv2.circle(img_copy, (x7, y7), size_points, verde, -1)
        cv2.circle(img_copy, (x8, y8), size_points, verde, -1)
        cv2.circle(img_copy, (x9, y9), size_points, verde, -1)

        # pierna izquierda
        x10 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].x * width)
        y10 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].y * height)
        x11 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_KNEE].x * width)
        y11 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_KNEE].y * height)
        x12 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ANKLE].x * width)
        y12 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ANKLE].y * height)
        cv2.line(img_copy, (x10, y10), (x11, y11), (255, 255, 255), size_points)
        cv2.line(img_copy, (x11, y11), (x12, y12), (255, 255, 255), size_points)
        cv2.circle(img_copy, (x10, y10), size_points, rosa, -1)
        cv2.circle(img_copy, (x11, y11), size_points, rosa, -1)
        cv2.circle(img_copy, (x12, y12), size_points, rosa, -1)

        # torso
        cv2.line(img_copy, (x10, y10), (x7, y7), blanco, 3)
        cv2.line(img_copy, (x1, y1), (x4, y4), blanco, 3)
        cv2.line(img_copy, (x4, y4), (x10, y10), blanco, 3)
        cv2.line(img_copy, (x1, y1), (x7, y7), blanco, 3)

        # pie derecho
        x13 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HEEL].x * width)
        y13 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HEEL].y * height)
        x14 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_FOOT_INDEX].x * width)
        y14 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_FOOT_INDEX].y * height)
        cv2.line(img_copy, (x13, y13), (x14, y14), (255, 255, 255), size_points)
        cv2.line(img_copy, (x13, y13), (x9, y9), (255, 255, 255), size_points)
        cv2.line(img_copy, (x14, y14), (x9, y9), (255, 255, 255), size_points)
        cv2.circle(img_copy, (x13, y13), size_points, verde, -1)
        cv2.circle(img_copy, (x14, y14), size_points, verde, -1)

        # pie izquierdo
        x15 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HEEL].x * width)
        y15 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HEEL].y * height)
        x16 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_FOOT_INDEX].x * width)
        y16 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_FOOT_INDEX].y * height)
        cv2.line(img_copy, (x15, y15), (x16, y16), (255, 255, 255), size_points)
        cv2.line(img_copy, (x15, y15), (x12, y12), (255, 255, 255), size_points)
        cv2.line(img_copy, (x16, y16), (x12, y12), (255, 255, 255), size_points)
        cv2.circle(img_copy, (x15, y15), size_points, rosa, -1)
        cv2.circle(img_copy, (x16, y16), size_points, rosa, -1)

        # mano izquierda
        x17 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_INDEX].x * width)
        y17 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_INDEX].y * height)
        x18 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_PINKY].x * width)
        y18 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_PINKY].y * height)
        x19 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_THUMB].x * width)
        y19 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_THUMB].y * height)
        cv2.line(img_copy, (x17, y17), (x18, y18), (255, 255, 255), size_points)
        cv2.line(img_copy, (x17, y17), (x6, y6), (255, 255, 255), size_points)
        cv2.line(img_copy,  (x18, y18), (x6, y6), (255, 255, 255), size_points)
        cv2.line(img_copy,  (x19, y19), (x6, y6), (255, 255, 255), size_points)
        cv2.circle(img_copy, (x17, y17), size_points, morado, -1)
        cv2.circle(img_copy, (x18, y18), size_points, morado, -1)
        cv2.circle(img_copy, (x19, y19), size_points, morado, -1)
     
        # mano derecha
        x20 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_INDEX].x * width)
        y20 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_INDEX].y * height)
        x21 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_PINKY].x * width)
        y21 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_PINKY].y * height)
        x22 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_THUMB].x * width)
        y22 = int(pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_THUMB].y * height)
        cv2.line(img_copy, (x20, y20), (x21, y21), (255, 255, 255), size_points)
        cv2.line(img_copy, (x20, y20), (x3, y3), (255, 255, 255), size_points)
        cv2.line(img_copy,  (x21, y21), (x3, y3), (255, 255, 255), size_points)
        cv2.line(img_copy,  (x20, y20), (x3, y3), (255, 255, 255), size_points)
        cv2.circle(img_copy, (x20, y20), size_points, naranja, -1)
        cv2.circle(img_copy, (x21, y21), size_points, naranja, -1)
        cv2.circle(img_copy, (x22, y22), size_points, naranja, -1)
    return img_copy
```

# Extraer los frames del video

```python
img_height = 192
img_width = 192

def crop_center_square(frame):
    y, x = frame.shape[0:2]
    min_dim = min(y, x)
    start_x = (x // 2) - (min_dim // 2)
    start_y = (y // 2) - (min_dim // 2)
    return frame[start_y:start_y+min_dim,start_x:start_x+min_dim]  

def load_video(path, resize=(img_height, img_width)):
    cap = cv2.VideoCapture(path)
    frames = []
    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            frame = crop_center_square(frame)
            frame = cv2.resize(frame, resize)
            frame = frame[:, :, [2, 1, 0]]
            frames.append(frame)
    finally:
        cap.release()
        
    return np.array(frames) / 255.0
```

```python
path_video = '../videos/ballet1.mp4'
frames = load_video(path_video)
```

# Estimar la pose

```python
def estimar_pose(frames):
    frames_pose = []
    for frame in frames:
        # Convert from float32 to uint8  
        frame = frame = (frame * 255).round().astype(np.uint8)
        # Covert BGR to RGB
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        # Process
        result = pose.process(frame)
        # Get pose landmarks
        pose_landmarks = result.pose_landmarks
        # Output frame with landmarks
        output_frame = drawPose2(frame,pose_landmarks)
        #output_frame = drawPose(frame,pose_landmarks)
        frames_pose.append(output_frame)
    return frames_pose
```

# Crear video apartir de frames

```python
def create_video(output_frames,fps=30,output_path='../videos/balle1_pose.mp4',frameSize = (192, 192)):
    out = cv2.VideoWriter(output_path,cv2.VideoWriter_fourcc(*'DIVX'), fps, frameSize)
    for i in range(len(output_frames)):
        out.write(output_frames[i])
    out.release()
```
