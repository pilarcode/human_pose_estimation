```python
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt
```

```python
mp_pose = mp.solutions.pose
```

```python
# Definicion del modelo
pose = mp_pose.Pose(static_image_mode=True,        
                              min_detection_confidence=0.3,
                              model_complexity=2)
```

```python
def drawPose(img, results,title="Detected Pose"):
    size_points = 25
    naranja =(128, 0, 255)
    morado = (255, 191, 0)
    verde = (6, 243, 161)
    amarillo = (236, 243, 6) 
    rosa = (224, 18, 145)
    azul = (18, 132, 224)
    negro = (0,0,0)
    blanco = (255, 255, 255)

     
    img_copy = img.copy()
    height, width, _ = img_copy.shape
    
    if results.pose_landmarks:

        #style=mp.solutions.drawing_styles.get_default_pose_landmarks_style()
        #mp.solutions.drawing_utils.draw_landmarks(
        #            image=img_copy,
        #            landmark_list=results.pose_landmarks,
        #           landmark_drawing_spec=style,                                        
        #            connections=mp.solutions.pose.POSE_CONNECTIONS)

        # brazo derecho
        x1 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].x * width)
        y1 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y * height)
        x2 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].x * width)
        y2 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].y * height)
        x3 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].x * width)
        y3 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].y * height)
        cv2.line(img_copy, (x1, y1), (x2, y2), (255, 255, 255), 3)
        cv2.line(img_copy, (x2, y2), (x3, y3), (255, 255, 255), 3)
        cv2.circle(img_copy, (x1, y1), size_points,naranja , -1)
        cv2.circle(img_copy, (x2, y2), size_points, naranja, -1)
        cv2.circle(img_copy, (x3, y3), size_points, naranja, -1)

        # brazo izquierdo
        x4 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x * width)
        y4 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y * height)
        x5 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].x * width)
        y5 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].y * height)
        x6 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].x * width)
        y6 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].y * height)
        cv2.line(img_copy, (x4, y4), (x5, y5), (255, 255, 255), 3)
        cv2.line(img_copy, (x5, y5), (x6, y6), (255, 255, 255), 3)
        cv2.circle(img_copy, (x4, y4), size_points, morado, -1)
        cv2.circle(img_copy, (x5, y5), size_points, morado, -1)
        cv2.circle(img_copy, (x6, y6), size_points, morado, -1)

        # pierna derecha
        x7 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].x * width)
        y7 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].y * height)
        x8 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].x * width)
        y8 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].y * height)
        x9 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ANKLE].x * width)
        y9 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ANKLE].y * height)
        cv2.line(img_copy, (x7, y7), (x8, y8), (255, 255, 255), 3)
        cv2.line(img_copy, (x8, y8), (x9, y9), (255, 255, 255), 3)
        cv2.circle(img_copy, (x7, y7), size_points, verde, -1)
        cv2.circle(img_copy, (x8, y8), size_points, verde, -1)
        cv2.circle(img_copy, (x9, y9), size_points, verde, -1)

        # pierna izquierda
        x10 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].x * width)
        y10 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].y * height)
        x11 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_KNEE].x * width)
        y11 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_KNEE].y * height)
        x12 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ANKLE].x * width)
        y12 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ANKLE].y * height)
        cv2.line(img_copy, (x10, y10), (x11, y11), (255, 255, 255), 3)
        cv2.line(img_copy, (x11, y11), (x12, y12), (255, 255, 255), 3)
        cv2.circle(img_copy, (x10, y10), size_points, rosa, -1)
        cv2.circle(img_copy, (x11, y11), size_points, rosa, -1)
        cv2.circle(img_copy, (x12, y12), size_points, rosa, -1)

        # torso
        cv2.line(img_copy, (x10, y10), (x7, y7), blanco, 3)
        cv2.line(img_copy, (x1, y1), (x4, y4), blanco, 3)
        cv2.line(img_copy, (x4, y4), (x10, y10), blanco, 3)
        cv2.line(img_copy, (x1, y1), (x7, y7), blanco, 3)

        # pie derecho
        x13 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HEEL].x * width)
        y13 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HEEL].y * height)
        x14 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_FOOT_INDEX].x * width)
        y14 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_FOOT_INDEX].y * height)
        cv2.line(img_copy, (x13, y13), (x14, y14), (255, 255, 255), 3)
        cv2.line(img_copy, (x13, y13), (x9, y9), (255, 255, 255), 3)
        cv2.line(img_copy, (x14, y14), (x9, y9), (255, 255, 255), 3)
        cv2.circle(img_copy, (x13, y13), size_points, blanco, -1)
        cv2.circle(img_copy, (x14, y14), size_points, blanco, -1)

        # pie izquierdo
        x15 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HEEL].x * width)
        y15 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HEEL].y * height)
        x16 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_FOOT_INDEX].x * width)
        y16 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_FOOT_INDEX].y * height)
        cv2.line(img_copy, (x15, y15), (x16, y16), (255, 255, 255), 3)
        cv2.line(img_copy, (x15, y15), (x12, y12), (255, 255, 255), 3)
        cv2.line(img_copy, (x16, y16), (x12, y12), (255, 255, 255), 3)
        cv2.circle(img_copy, (x15, y15), size_points, blanco, -1)
        cv2.circle(img_copy, (x16, y16), size_points, blanco, -1)

        # mano izquierda
        x17 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_INDEX].x * width)
        y17 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_INDEX].y * height)
        x18 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_PINKY].x * width)
        y18 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_PINKY].y * height)
        x19 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_THUMB].x * width)
        y19 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_THUMB].y * height)
        cv2.line(img_copy, (x17, y17), (x18, y18), (255, 255, 255), 3)
        cv2.line(img_copy, (x17, y17), (x6, y6), (255, 255, 255), 3)
        cv2.line(img_copy,  (x18, y18), (x6, y6), (255, 255, 255), 3)
        cv2.line(img_copy,  (x19, y19), (x6, y6), (255, 255, 255), 3)
        cv2.circle(img_copy, (x17, y17), size_points, blanco, -1)
        cv2.circle(img_copy, (x18, y18), size_points, blanco, -1)
        cv2.circle(img_copy, (x19, y19), size_points, blanco, -1)
     
        # mano derecha
        x20 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_INDEX].x * width)
        y20 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_INDEX].y * height)
        x21 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_PINKY].x * width)
        y21 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_PINKY].y * height)
        x22 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_THUMB].x * width)
        y22 = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_THUMB].y * height)
        cv2.line(img_copy, (x20, y20), (x21, y21), (255, 255, 255), 3)
        cv2.line(img_copy, (x20, y20), (x3, y3), (255, 255, 255), 3)
        cv2.line(img_copy,  (x21, y21), (x3, y3), (255, 255, 255), 3)
        cv2.line(img_copy,  (x20, y20), (x3, y3), (255, 255, 255), 3)
        cv2.circle(img_copy, (x20, y20), size_points, blanco, -1)
        cv2.circle(img_copy, (x21, y21), size_points, blanco, -1)
        cv2.circle(img_copy, (x20, y20), size_points, blanco, -1)

    img_copy = cv2.cvtColor(img_copy, cv2.COLOR_RGB2BGR)
    fig = plt.figure(figsize = [10, 10])
    plt.title(title)
    plt.axis('off')
    plt.imshow(img_copy[:,:,::-1])
    plt.show()
```

```python
img1 = cv2.imread('..\images\pose5.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
results1 = pose.process(img1)
drawPose(img1,results1)
```
