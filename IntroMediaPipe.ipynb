```python
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt
```

```python
# Definicion del modelo
pose = mp.solutions.pose.Pose(static_image_mode=True,        
                              min_detection_confidence=0.3,
                              model_complexity=2)
```

```python
def drawPose(img, results,title="Detected Pose"):
 img_copy = img.copy()

 if results.pose_landmarks:

  style=mp.solutions.drawing_styles.get_default_pose_landmarks_style()
  mp.solutions.drawing_utils.draw_landmarks(
                   image=img_copy,
                   landmark_list=results.pose_landmarks,
                   landmark_drawing_spec=style,                                        
                   connections=mp.solutions.pose.POSE_CONNECTIONS)

  fig = plt.figure(figsize = [10, 10])
  plt.title(title)
  plt.axis('off')
  plt.imshow(img_copy[:,:,::-1])
  plt.show()
```

```python
img1 = cv2.imread('images\pose1.jpg')
results1 = pose.process(img1)
drawPose(img1,results1,"")
```

```python
img2 = cv2.imread('images\pose3.jpg')
results2 = pose.process(img2)
drawPose(img2,results2,"")
```

```python
img3 = cv2.imread('images\pose4.jpg')
results3 = pose.process(img3)
drawPose(img3,results3,"")
```

```python
image = cv2.imread('images\pose5.jpg') 
results = pose.process(image)
drawPose(image,results,"")
```
